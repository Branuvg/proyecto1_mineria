```python
from mlxtend.frequent_patterns import apriori, association_rules
import sys
```

## AnÃ¡lisis Sugeridos con MLxtend y Sklearn

### **Association Rules (Apriori)** - Para variables categÃ³ricas
- **QuÃ© analizar:** Reglas de asociaciÃ³n entre edad, escolaridad, departamento, mes, etc.
- **Ãštil para:** Encontrar patrones ocultos de matrimonios/divorcios

### **Clustering (KMeans, DBSCAN)** - Agrupar perfiles similares
- **QuÃ© analizar:** Agrupar matrimonios/divorcios por caracterÃ­sticas similares
- **Ãštil para:** SegmentaciÃ³n y perfiles demogrÃ¡ficos

### **Correlaciones (Spearman/Pearson)** - Variables numÃ©ricas
- **QuÃ© analizar:** RelaciÃ³n entre edad_hombre, edad_mujer, mes, aÃ±o
- **Ãštil para:** Entender relaciones lineales

---
## Ejemplo 1: Correlaciones entre variables numÃ©ricas

```python
# 1. CORRELACIÃ“N entre variables numÃ©ricas (matrimonios)
# Seleccionar solo columnas numÃ©ricas de las columnas comunes
numeric_cols = ['EDADHOM', 'EDADMUJ', 'ESCHOM', 'ESCMUJ', 'MESREG', 'MESOCU']
matrs_numeric = matrs[numeric_cols].dropna()

# Calcular matriz de correlaciÃ³n
correlation_matrix = matrs_numeric.corr()

print(" MATRIZ DE CORRELACIÃ“N - MATRIMONIOS\n")
print("="*70)
print(correlation_matrix.round(3))
print("="*70)

print("\n Correlaciones mÃ¡s fuertes:")
# Encontrar las correlaciones mÃ¡s fuertes (excluyendo diagonal)
corr_pairs = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        corr_pairs.append((
            correlation_matrix.columns[i],
            correlation_matrix.columns[j],
            correlation_matrix.iloc[i, j]
        ))

# Ordenar por valor absoluto
corr_pairs.sort(key=lambda x: abs(x[2]), reverse=True)

for var1, var2, corr in corr_pairs[:10]:
    if abs(corr) > 0.1:  # Solo mostrar correlaciones significativas
        emoji = "ðŸ”´" if abs(corr) > 0.5 else "ðŸŸ¡" if abs(corr) > 0.2 else "ðŸŸ¢"
        print(f"{emoji} {var1} â†” {var2}: {corr:.3f}")
```

## Ejemplo 2: Association Rules (Apriori) - Reglas de asociaciÃ³n

```python
# 2. ASSOCIATION RULES - Discretizar variables y buscar patrones
# Preparar datos: discretizar variables continuas en categorÃ­as

matrs_sample = matrs.sample(n=min(10000, len(matrs)), random_state=42).copy()

# Discretizar edades en rangos
matrs_sample['edad_hom_rango'] = pd.cut(matrs_sample['EDADHOM'], 
                                         bins=[0, 25, 35, 45, 100], 
                                         labels=['18-25', '26-35', '36-45', '46+'])

matrs_sample['edad_muj_rango'] = pd.cut(matrs_sample['EDADMUJ'], 
                                         bins=[0, 25, 35, 45, 100], 
                                         labels=['18-25', '26-35', '36-45', '46+'])

# Crear variables categÃ³ricas
matrs_sample['escolaridad_hom'] = matrs_sample['ESCHOM'].map({
    1: 'Ninguno', 2: 'Primaria', 3: 'Secundaria', 
    4: 'Diversificado', 5: 'Universidad', 6: 'Posgrado', 9: 'No_esp'
})

matrs_sample['mes_registro'] = matrs_sample['MESREG'].map({
    1: 'Ene', 2: 'Feb', 3: 'Mar', 4: 'Abr', 5: 'May', 6: 'Jun',
    7: 'Jul', 8: 'Ago', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dic'
})

# Convertir a formato one-hot para Apriori
categorical_cols = ['edad_hom_rango', 'edad_muj_rango', 'escolaridad_hom', 'mes_registro']
matrs_encoded = pd.get_dummies(matrs_sample[categorical_cols].dropna())

# Aplicar Apriori
frequent_itemsets = apriori(matrs_encoded, min_support=0.05, use_colnames=True)

# Generar reglas de asociaciÃ³n
if len(frequent_itemsets) > 0:
    rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
    rules = rules.sort_values('lift', ascending=False)
    
    print(f" Se encontraron {len(rules)} reglas de asociaciÃ³n\n")
    print(" Top 10 reglas con mayor lift:\n")
    
    for idx, rule in rules.head(10).iterrows():
        antecedents = ', '.join(list(rule['antecedents']))
        consequents = ', '.join(list(rule['consequents']))
        print(f" SI [{antecedents}]")
        print(f"   â†’ ENTONCES [{consequents}]")
        print(f"   â€¢ Confianza: {rule['confidence']:.2%}")
        print(f"   â€¢ Lift: {rule['lift']:.2f}")
        print(f"   â€¢ Soporte: {rule['support']:.2%}\n")
else:
    print(" No se encontraron itemsets frecuentes. Intenta reducir min_support.")
```

---
## Ejemplo 3: Clustering (KMeans) - SegmentaciÃ³n de perfiles

```python
# 3. CLUSTERING - Agrupar matrimonios por caracterÃ­sticas similares

# Preparar datos para clustering
cluster_features = ['EDADHOM', 'EDADMUJ', 'ESCHOM', 'ESCMUJ', 'DEPREG']
matrs_cluster = matrs[cluster_features].dropna()

# Tomar muestra para performance
matrs_cluster_sample = matrs_cluster.sample(n=min(5000, len(matrs_cluster)), random_state=42)

# Normalizar datos
scaler = StandardScaler()
matrs_scaled = scaler.fit_transform(matrs_cluster_sample)

# MÃ©todo del codo para encontrar K Ã³ptimo
inertias = []
K_range = range(2, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(matrs_scaled)
    inertias.append(kmeans.inertia_)

# Mostrar inercias (mÃ©todo del codo en texto)
print(" MÃ‰TODO DEL CODO - Valores de Inercia (WCSS)")
print("="*50)
for k, inertia in zip(K_range, inertias):
    print(f"K={k:2d} â†’ Inercia: {inertia:,.2f}")
print("="*50)
print(" Busca el 'codo' donde la inercia deja de disminuir drÃ¡sticamente")

# Aplicar KMeans con K=4 (puedes ajustar segÃºn el grÃ¡fico)
k_optimo = 4
kmeans_final = KMeans(n_clusters=k_optimo, random_state=42, n_init=10)
clusters = kmeans_final.fit_predict(matrs_scaled)

# Agregar clusters al dataframe
matrs_cluster_sample['Cluster'] = clusters

# Analizar caracterÃ­sticas de cada cluster
print(f" AnÃ¡lisis de {k_optimo} clusters identificados:\n")

for i in range(k_optimo):
    cluster_data = matrs_cluster_sample[matrs_cluster_sample['Cluster'] == i]
    print(f"{'='*60}")
    print(f"CLUSTER {i} (n={len(cluster_data)} matrimonios)")
    print(f"{'='*60}")
    print(f"  Edad Hombre: {cluster_data['EDADHOM'].mean():.1f} aÃ±os (Â±{cluster_data['EDADHOM'].std():.1f})")
    print(f"  Edad Mujer:  {cluster_data['EDADMUJ'].mean():.1f} aÃ±os (Â±{cluster_data['EDADMUJ'].std():.1f})")
    print(f"  Escolaridad Hombre: {cluster_data['ESCHOM'].mean():.1f}")
    print(f"  Escolaridad Mujer:  {cluster_data['ESCMUJ'].mean():.1f}")
    print(f"  Departamento mÃ¡s comÃºn: {cluster_data['DEPREG'].mode().values[0]}")
    print()

# Calcular silhouette score
silhouette_avg = silhouette_score(matrs_scaled, clusters)
print(f" Silhouette Score: {silhouette_avg:.3f}")
print("   (Cercano a 1 = muy buenos clusters, cercano a 0 = clusters solapados)")
```
